```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Original data with None values
data = {
    'User': ['Sam and Bishop', 'Robert and Bob', 'Cody', 'Alphonso', 'Leonel'],
    'Dune: Part Two': [None, 3.0, None, 4.0, 3.0],
    'Nosferatu': [None, 5.0, 2.0, None, 3.0],
    'The Beekeeper': [5.0, None, None, 3.0, 4.0],
    'Inside Out 2': [None, None, 5.0, 4.0, 2.0],
    'DeadPool and Wolverine': [5.0, 4.0, None, 2.0, None],
    'The Monkey': [None, None, None, None, 4.0]
}

df = pd.DataFrame(data)

# Replace None with 0
df_filled_with_zeros = df.fillna(0)

# Normalize the ratings
df_normalized_zeros = df_filled_with_zeros.drop('User', axis=1).apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=0)
df_normalized_zeros['User'] = df['User']

# Visualization: Heatmap with 0 filled ratings
plt.figure(figsize=(10, 6))  
sns.heatmap(df_normalized_zeros.set_index('User').T, annot=True, cmap="coolwarm", cbar=True)
plt.title('Normalized Movie Ratings Heatmap (Filled with 0)')
plt.show()

# Display the normalized ratings for inspection
print("Normalized Ratings with 0 filled in:")
print(df_normalized_zeros)
```

```python
In this notebook, I will analyze movie ratings given by a group of users. The ratings are normalized and visualized using a heatmap to understand the differences in preferences.

I chose to fill the missing values with the column average to avoid biasing the analysis with zero ratings.

The heatmap shows the normalized ratings for each user across the six movies. User Cody rated 'Inside Out 2' very highly, while others gave it lower scores.

This analysis shows that User Cody has a very specific taste for animated movies, while other users have more varied preferences. Normalizing the ratings helped mitigate the differences in the overall rating scale.

My source for the movies is https://www.boxofficemojo.com/year/world/2025/ and https://www.boxofficemojo.com/year/world/2024/

```
